```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('Comprehensive_Banking_Database.csv')

dates=df["Date Of Account Opening"]

d=np.arange(2000,2023)
print(d)
l=[0]*23
for date in dates:
    year=int(date.split('/')[2])
    l[year%2000]+=1
print(l)

plt.figure(figsize=(10,5))
plt.plot(d,l,marker='o',ms=3,color='b')
plt.xlabel('Year')
plt.ylabel('No of Account')
plt.title("New Account openings per year")
plt.show()
```
