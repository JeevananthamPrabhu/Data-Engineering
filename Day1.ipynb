```python
m=[
    [1,2,3,11],
    [4,5,6,10],
    [7,8,9,12]
]
tm=[]
for i in range(len(m[0])):
    t=[]
    for j in range(len(m)) :
        t.append(m[j][i])
    tm.append(t)
print(tm)      
```

```python
tm=[[m[j][i] for j in range(len(m))] for i in range(len(m[0]))]
print(tm)
```

```python
#set operation
l1={1,2,3,4,5,6}
l2={4,5,6,7,8}
print(l1 | l2)
print(l1.union(l2))
print(l1 & l2)
print(l1.intersection(l2))
print(l1 - l2)
print(l1.difference(l2))
print(l1 ^ l2)
print(l1.symmetric_difference(l2))
```
