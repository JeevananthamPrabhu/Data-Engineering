```python
df=pd.read_csv('blood_pressure.csv')
df
```

```python
age_group=df['agegrp'].unique()
age_group
```

```python
before=[]
after=[]
for i in age_group:
    b=df.loc[df['agegrp']==i,['bp_before']]
    a=df.loc[df['agegrp']==i,['bp_after']]
    before.append(b.mean()['bp_before'])
    after.append(a.mean()['bp_after'])
print(before)
print(after)
```

```python
plt.plot(age_group,before,marker='o',color='r',label='before')
plt.plot(age_group,after,marker='s',color='g',label='after')
plt.xlabel('Age Group')
plt.ylabel('Blood Pressure Level')
plt.legend()
plt.show()
```

```python
cat=['A','B','C']
val=[15,40,30]
cnt1=plt.bar(cat,val,color=['r','g','b'],width=0.5)
plt.bar_label(cnt1)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
cat=['A','B','C']
val=[15000,40000,30000]
cnt1=plt.bar(cat,val,color=['r','g','b'],width=0.5)
plt.bar_label(cnt1,fmt='{:,.0f}',label_type='center')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```
