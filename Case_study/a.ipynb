# Gender Distribution Analysis

#### Count number of males and females

```python
total=len(df)
males=len(df[df['sex']==1])
females=len(df[df['sex']==0])
print("Total : ",total)
print("Males : ",males)
print("Females : ",females)
```

#### Calculate percentage distribution using NumPy

```python

males_per = (males / total) * 100
females_per = (females / total) * 100

print(f"Males percentage: {males_per:.2f}%")
print(f"Females percentage: {females_per:.2f}%")
```

#### Plot a bar chart using Matplotlib

```python
cnt=plt.bar(['Males','Females'],[males,females],width=0.5,color='b')
plt.bar_label(cnt,label_type='center')
plt.title('No of Patients by gender')
plt.xlabel('Gender')
plt.ylabel('No of patient')
plt.show()
```

# Create Risk Score (Custom Analysis)

#### Create new column:
#### risk_score = (chol/200) + (trestbps/120) + (oldpeak)
#### Classify patients as:
#### Low Risk
#### Medium Risk
#### High Risk
#### Visualize distribution

```python

df['risk_score'] = (df['chol'] / 200) + (df['trestbps'] / 120) + df['oldpeak']

conditions = [
    (df['risk_score'] < 2),
    (df['risk_score'] >= 2) & (df['risk_score'] < 3),
    (df['risk_score'] >= 3)
]
choices = ['Low Risk', 'Medium Risk', 'High Risk']
df['risk_class'] = np.select(conditions, choices, default='Low Risk')


plt.figure(figsize=(8, 6))
df['risk_score'].plot(kind='hist')
plt.title('Risk Score Distribution')
plt.xlabel('Risk Score')
plt.ylabel('Frequency')
plt.show()


df['risk_class'].value_counts().plot(kind='bar',color=['r','b','g'])
plt.title('Risk Class Distribution')
plt.xlabel('Risk Class')
plt.ylabel('Number of Patients')
plt.show()
```
