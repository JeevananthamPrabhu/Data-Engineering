```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('Comprehensive_Banking_Database.csv')

df.columns

feedback_type=df['Feedback Type'].unique()
```

```python
resolved=[]
pending=[]
for ty in feedback_type:
    temp=df[df['Feedback Type']==ty]
    resolved.append(len(temp[temp['Resolution Status']=='Resolved']))
    pending.append(len(temp[temp['Resolution Status']=='Pending']))
print(resolved)
print(pending)

x=np.arange(3)
cnt1=plt.bar(x-0.1, resolved,width=0.2, label='Resolved', color='g')
cnt2=plt.bar(x+0.1, pending,width=0.2, label='Pending', color='r')
plt.xticks(x,feedback_type)
plt.xlabel("Feedback Type")
plt.ylabel("No of Feedbacks")
plt.bar_label(cnt1)
plt.bar_label(cnt2)
plt.title("Feedback Status")
plt.legend()
plt.show()
```

```python
sug=resolved[0]+pending[0]
com=resolved[1]+pending[1]
pra=resolved[2]+pending[2]
print(sug,com,pra)

plt.pie([sug,com,pra],labels=feedback_type,autopct='%.2f%%')
plt.title("Feedback Types")
plt.legend()
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(12,6))

axs[0].pie([sug,com,pra],labels=feedback_type,autopct='%.2f%%')
axs[0].set_title("Feedback Types")

cnt1=axs[1].bar(x-0.1, resolved,width=0.2, label='Resolved', color='g')
cnt2=axs[1].bar(x+0.1, pending,width=0.2, label='Pending', color='r')
axs[1].bar_label(cnt1)
axs[1].bar_label(cnt2)
axs[1].set_xticks(x,feedback_type)
axs[1].set_xlabel("Feedback Type")
axs[1].set_ylabel("No of Feedbacks")
axs[1].set_title("Feedback Status")
axs[1].legend()
plt.show()
```
