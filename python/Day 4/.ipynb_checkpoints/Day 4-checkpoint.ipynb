# Iterators

An iterator is object that allows to traverse

```python
class Nums:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<10:
            self.num+=1
            return self.num-1
        else:
            raise StopIteration
n1=Nums()
it=iter(n1)
for i in it:
    print(i)

```

```python
class Even:
    def __iter__(self):
        self.num=0
        return self
    def __next__(self):
        if self.num<=10:
            self.num=+2
            return self.num-2
        else:
            raise StopIteration
e1=Even()
for i in iter(e1):
    print(i)
```

```python
def my():
    for i in range(10):
        yield i
g=my()
for i in g:
    print(i)
```

#### iterator :
iter and next
more code, manual implementation
slightly more control
#### generator:
yield
simple,concise implementation
more memory efficient
