FUNCTION

```python
#create fuction by using keyword def
def sum(lst):
    '''
    return sum of all elements
    '''
    s=0
    for num in lst:
        s+=num
    return s
print(sum([1,2,3,4,5]))
#print(s)
print(sum.__doc__) #returns the doc string
```

```python
def hcf(num1,num2):
    l=num1 if num1>num2 else num2
    for i in range(1,l):
        if num1%i==0 and num2%i==0:
            ans=i
    return ans
def gcd(num1,num2):
    if num1==0:
        return num2
    return gcd(num2%num1,num1)
print(hcf(14,8))
print(gcd(14,8))      
```

IN BUILT FUCTION

```python
#enumerate
lst=[2,3,4,1,2]
for i,num in enumerate(lst): #default:start=0
    print("{}:{}".format(i,num))
```

```python
#change start to 1
for i,num in enumerate(lst,start=1):
    print("{}:{}".format(i,num))
```

```python
#change the postion of format by index
for i,num in enumerate(lst):
    print("{1}:{0}".format(i,num))
```

```python
#filter 
#extract the items
#pass the fuction that returns true or false
def posi(n):
    if n>0:
        return n
l=list(filter(posi,[1,-1,2,-2,-3,5]))
print(l)
#using lambda
l=list(filter(lambda n:n>0,[1,-1,2,-2,-3,5]))
print(l)
```

```python
#map
#maps a fuction to all items
def square(n):
    return n*n
sq=list(map(square,[1,2,3,4]))
print(sq)
#using lambda
sq=list(map(lambda n:n*n,[1,2,3,4]))
print(sq)
```

```python
#reduce
#computation on sequential pair 
#return single result
#import from functools
from functools import reduce
def mul(n1,n2):
    return n1*n2
print(reduce(mul,[1,2,3,4,5]))
#using lambda
print(reduce(lambda n1,n2:n1*n2,[1,2,3,4,5]))
```

ARGUMENTS

```python
def arg(a,b,c):
    print(a,b,c)
arg(1,2,3)
```

```python
#default argument
#default argument must at end
def arg(a,b,c=3): #c is get default value
    print(a,b,c)
arg(1,2) #c not passed
arg(1,2,5) #overrides
```

```python
#arbitary argument
#can pass multiple arguments
def arg(*a):
    print(a)
arg(2,'ram',20)
```

```python
#keyword argument
#cam pass multiple arguments with name
def arg(**k):
    print(k)
arg(id=20,name='ram')
```

```python
class Student:
    def __init__(self,id,name):
        self.id=id
        self.name=name
s1=Student(2,"ram")
s2=Student(3,"ravi")
print(s1.id,s1.name)
print(s2.id,s2.name)
```

```python
class Student:
    def __init__(self,id,name):
        self.id=id
        self.name=name
    def desc(self):
        print("id:{},name:{}".format(self.id,self.name))
s1=Student(5,"dj")
s1.desc()
```

INHERITANCE

```python
class A:
    id=1
    name="a"
class B(A):
    name="b"
b=B()
print(b.id,b.name)
```

```python
class Car:
    def __init__(self,model,colour):
        self.model=model
        self.colour=colour
class AcCar(Car):
    def __init__(self,model,colour,ac=False):
        super().__init__(model,colour)
        self.ac=ac
car1=AcCar("321","black",True)
print(car1.model,car1.colour,car1.ac)
        
```
