#### Why we cannot use local variable outside the function

```python
# Variables declare inside a function belong only to that function
# Once the function finishes executing, the variable is not accessible
def fun(l):
    t=0
    for i in l:
        t+=i
    return t
print(fun([1,2,3,4,5]))
# print(t) -> error variable not accessible
```

#### Where do you use map,reduce and filter

```python
#same operation to every item
l=['1','2','3']
print(list(map(int,l)))
```

```python
#combine all items into a single result
from functools import reduce
l=[1,2,3,4]
print(reduce(lambda x,y:x+y,l))
```

```python
#only keeps items that match with condition
l=[1,2,3,4]
print(list(filter(lambda x:x%2==0,l)))
```
