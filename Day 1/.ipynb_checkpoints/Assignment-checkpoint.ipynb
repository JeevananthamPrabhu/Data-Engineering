#### Matrix Transpose

```python
m=[
    [1,2,3,11],
    [4,5,6,10],
    [7,8,9,12]
]
tm=[]
#Transpose using nested for loop
for i in range(len(m[0])):
    t=[]
    for j in range(len(m)) :
        t.append(m[j][i])
    tm.append(t)
print(tm)      
```

```python
#Transpose using List Comprehension
tm=[[m[j][i] for j in range(len(m))] for i in range(len(m[0]))]
print(tm)
```

#### Assignment share elements without sharing address

```python
#Copy
l=[2,3]
l1=l #Makes refer to same address
l2=l.copy() #Creates the copy
l[0]=6
print(l1)
print(l2)
# #id -> shows address 
print(id(l))
print(id(l1))
print(id(l2)) #different
```
